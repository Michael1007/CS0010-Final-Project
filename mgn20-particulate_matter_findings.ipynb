```python
#import statments
import pandas as pd
import numpy as np
import geopandas
%matplotlib inline
import matplotlib.pyplot as plt
```

# Data Sheet

```python
#import dataset
partMat = pd.read_csv("pm2.52011ct.csv")
#filter the important info
num_PM = partMat.groupby("CensusTract").sum()['PM25']
num_PM.sort_values(ascending=True)
```

```python
#read in shape file
neighborhoods = geopandas.read_file("Particulate_Matter/ParticulateMatter2.5.shp")
#plot shape file
#neighborhoods.plot()
```

```python
#now to merge the data
pm_map = neighborhoods.merge(num_PM, how='left', left_on='PM25', right_on="PM25")
#look at head to confirm it merged correctly
#pm_map[['CensusTrac','PM25', 'geometry']].head()
```

```python
#draw a "choropleth" map using data from column
pm_map.plot(column='PM25', 
            cmap='magma',
            edgecolor="white", 
            legend = True, 
            legend_kwds={'label': "Mean PM 2.5 per cubic meter, in micrograms"}, 
            figsize = (15,10)
           )
```
